```python
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Path to your Excel file
ei_file_path = '/content/EI_chat_gpt.xlsx'

# Loading data from each sheet
ei_ita = pd.read_excel(ei_file_path, sheet_name='EI ITA')
ei_fr = pd.read_excel(ei_file_path, sheet_name='EI FR')
ei_ger = pd.read_excel(ei_file_path, sheet_name='EI GER')
```

```python
# Extracting EI scores
ei_scores_ita = ei_ita['Italy EI Individual Level']
ei_scores_fr = ei_fr['France EI Individual Level']
ei_scores_ger = ei_ger['Germany EI Invididual Level']

# T-tests
ttest_ita_fr = ttest_ind(ei_scores_ita, ei_scores_fr, equal_var=False)
ttest_ita_ger = ttest_ind(ei_scores_ita, ei_scores_ger, equal_var=False)
ttest_fr_ger = ttest_ind(ei_scores_fr, ei_scores_ger, equal_var=False)

print("Italy vs France:", ttest_ita_fr)
print("Italy vs Germany:", ttest_ita_ger)
print("France vs Germany:", ttest_fr_ger)
```

```python
# EI Data Plot
plt.figure(figsize=(15, 6))  # Adjust these values as needed

plt.subplot(1, 2, 2)
sns.boxplot(data=[ei_scores_ita,ei_scores_fr, ei_scores_ger])
plt.xticks([0, 1, 2], ['Italy', 'France', 'Germany'])  # Set country names for EI plot
plt.title('Boxplot of EI Data Variation', fontsize=17, fontweight='bold')
plt.ylabel('EI Level', fontsize=15)
```
