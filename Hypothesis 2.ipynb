```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Adjust the file path and sheet name as per your data
ma_file_path = '/content/M&A_chat_gpt.xlsx'
ma_data = pd.read_excel(ma_file_path)
```

```python
# Filtering M&A data by country
ma_data_ita = ma_data[ma_data['Acquirer Country'] == 'ITA']['Normalized SI']
ma_data_fr = ma_data[ma_data['Acquirer Country'] == 'FR']['Normalized SI']
ma_data_ger = ma_data[ma_data['Acquirer Country'] == 'GER']['Normalized SI']
```

```python
# France Acquiring Germany vs France Acquiring Italy
palette_colors = sns.color_palette("deep")


plt.figure(figsize=(10, 8))
ax = sns.boxplot(data=[ma_data_ita, ma_data_fr, ma_data_ger],  palette = [palette_colors[0], palette_colors[1], palette_colors[2]])

# Impostazione dei label dell'asse x
plt.xticks([0, 1, 2], ['Italy', 'France', 'Germany'])
plt.title('Boxplot of Normalized SI by Country', fontsize=16, fontweight='bold')
plt.ylabel('Normalized SI',  fontsize=15)

```
