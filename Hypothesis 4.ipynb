```python
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Adjust the file path and sheet name as per your data
ma_file_path = '/content/M&A_chat_gpt.xlsx'
ma_data = pd.read_excel(ma_file_path)
```

```python
fr_acquiring_ger = ma_data[(ma_data['Acquirer Country'] == 'FR') & (ma_data['Target Country'] == 'GER')]['Normalized SI']
fr_acquiring_it = ma_data[(ma_data['Acquirer Country'] == 'FR') & (ma_data['Target Country'] == 'ITA')]['Normalized SI']
it_acquiring_ger = ma_data[(ma_data['Acquirer Country'] == 'ITA') & (ma_data['Target Country'] == 'GER')]['Normalized SI']
it_acquiring_fr = ma_data[(ma_data['Acquirer Country'] == 'ITA') & (ma_data['Target Country'] == 'FR')]['Normalized SI']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))


# France Acquiring Germany vs France Acquiring Italy
palette_colors = sns.color_palette("deep")


plt.subplot(1, 2, 1)
sns.boxplot(data=[fr_acquiring_ger, fr_acquiring_it], palette = [palette_colors[1], palette_colors[1]])
plt.xticks([0, 1], ['FR Acquiring GER', 'FR Acquiring ITA'])
plt.title('France Acquisitions SI Comparison', fontsize=16, fontweight='bold')
plt.ylabel('Normalized SI', fontsize=15)

# Italy Acquiring Germany vs Italy Acquiring France
plt.subplot(1, 2, 2)
sns.boxplot(data=[it_acquiring_ger, it_acquiring_fr], palette = [palette_colors[0], palette_colors[0]])
plt.xticks([0, 1], ['ITA Acquiring GER', 'ITA Acquiring FR'])
plt.title('Italy Acquisitions SI Comparison', fontsize=16, fontweight='bold')
plt.ylabel('Normalized SI', fontsize=15)

plt.tight_layout()
plt.show()
```
